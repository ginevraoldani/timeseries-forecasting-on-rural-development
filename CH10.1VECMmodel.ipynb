# VECM model for cointegrated groups of variables

[var vs vecm - medium](https://medium.com/@uradeneha/unveiling-the-dynamics-a-comprehensive-guide-to-var-vecm-models-in-time-series-analysis-e95de1dfddb2) <br>
Vector Error Correction Model is an extension of the VAR model that accounts for both short-term deviations and long-term relationships among multiple time series variables. VECM is particularly useful when dealing with non-stationary time series data and aims to correct for any cointegration among the variables.

VECM introduces an error correction term, which represents the adjustment mechanism that brings the variables back to their long-term equilibrium relationship.
Cointegration, a condition where linear combinations of non-stationary variables are stationary, is a prerequisite for using VECM.
VECM is commonly applied in the presence of non-stationary variables to capture both short-run and long-run dynamics simultaneously.

Note: VAR models provide a framework for modelling the interdependencies among multiple time series variables, while VECM extends this concept to address issues related to non-stationarity and cointegration. Both models are valuable tools in time series forecasting, offering insights into the complex dynamics of interconnected variables over time.
